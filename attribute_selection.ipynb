```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import chi2, SelectKBest
```

```python
from google.colab import drive
drive.mount("/content/drive")
```

```python
ad = pd.read_csv("/content/drive/MyDrive/curdo de ai/CursoIA/4.Tópicos Avançados de Machine Learning/ad.data", header=None)
ad.shape
```

```python
x = ad.iloc[:,:-1].values
y = ad.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
model1 = GaussianNB()
model1.fit(x_train, y_train)
predict1 = model1.predict(x_test)
accuracy_score(y_test, predict1)
```

```python
selection = SelectKBest(chi2, k=7)
x_new = selection.fit_transform(x, y)
```

```python
x_new
```

```python
x_new.shape
```

```python
print(selection.get_support())
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_new, y, test_size=0.3, random_state=0)
```

```python
model2 = GaussianNB()
model2.fit(x_train, y_train)
predict2 = model2.predict(x_test)
accuracy_score(y_test, predict2)
```
