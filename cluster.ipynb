```python
iris = datasets.load_iris()
cluster1 = scale(iris.data)
cluster2 = scale(np.random.rand(150, 4))
```

```python
inertia = []

for i in range(1, 8):
  k_means = KMeans(n_clusters=i, n_init="auto")
  k_means.fit(cluster1)
  inertia.append(k_means.inertia_)

plt.plot(range(1, 8), inertia)
plt.title("Elbow")
plt.show()
```

```python
inertia = []

for i in range(1, 8):
  k_means = KMeans(n_clusters=i, n_init="auto")
  k_means.fit(cluster2)
  inertia.append(k_means.inertia_)

plt.plot(range(1, 8), inertia)
plt.title("Elbow")
plt.show()
```
