```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTENC
```

```python
credit = pd.read_csv("/content/drive/MyDrive/credit_simple.csv", sep = ";")
```

```python
credit.head()
```

```python
count = credit.groupby(['CLASSE']).size()
count
```

```python
y = credit['CLASSE'].values
x = credit.iloc[:,:-1].values
```

```python
label_enconder = LabelEncoder()

for i in range(x.shape[1]):
  if x[:,i].dtype == 'object':
    x[:,i] = label_enconder.fit_transform(x[:,i])

sm = SMOTENC(random_state=0, categorical_features=[3, 5, 6])

x_res, y_res = sm.fit_resample(x, y)
```

```python
df = pd.DataFrame({'CLASSE': y_res})
df.value_counts()
```
